# Data Pre-processing

Frist of all I will handel the unblanced data to make sure about the final model accuracy 

### Oversampling : 

```python
from imblearn.over_sampling import SMOTE
```

```python
oversampling=SMOTE()
X,y=oversampling.fit_resample(data.drop(['Class'], axis = 1),data['Class'])
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

scalar = StandardScaler()

#X = data.drop('Class', axis=1)
#y = data.Class

X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.3, random_state=42)
#X_train, X_validate, y_train, y_validate = train_test_split(X_train_v, y_train_v, 
                                                          # old way test_size=0.2, random_state=42)

X_train_scaled = scalar.fit_transform(X_train)
#X_validate = scalar.transform(X_validate)
X_test_scaled = scalar.transform(X_test)

w_p = y_train.value_counts()[0] / len(y_train)
w_n = y_train.value_counts()[1] / len(y_train)

print(f"Fraudulant transaction weight: {w_n}")
print(f"Non-Fraudulant transaction weight: {w_p}")
```

```python
print(f"TRAINING: X_train: {X_train.shape}, y_train: {y_train.shape}\n{'_'*55}")
#print(f"VALIDATION: X_validate: {X_validate.shape}, y_validate: {y_validate.shape}\n{'_'*50}")
print(f"TESTING: X_test: {X_test.shape} y_test: {y_test.shape}")
```

# Model Bulding 


```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, f1_score

def print_score(label, prediction, train=True):
    if train:
        clf_report = pd.DataFrame(classification_report(label, prediction, output_dict=True))
        print("Train Result:\n================================================")
        print(f"Accuracy Score: {accuracy_score(label, prediction) * 100:.2f}%")
        print("_______________________________________________")
        print(f"Classification Report:\n{clf_report}")
        print("_______________________________________________")
        print(f"Confusion Matrix: \n {confusion_matrix(y_train, prediction)}\n")
        
    elif train==False:
        clf_report = pd.DataFrame(classification_report(label, prediction, output_dict=True))
        print("Test Result:\n================================================")        
        print(f"Accuracy Score: {accuracy_score(label, prediction) * 100:.2f}%")
        print("_______________________________________________")
        print(f"Classification Report:\n{clf_report}")
        print("_______________________________________________")
        print(f"Confusion Matrix: \n {confusion_matrix(label, prediction)}\n") 
```

### 1- Logistic Regression

```python
# modeling imports
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn import metrics
from sklearn.model_selection import KFold
import numpy as np


X_scaled = scalar.fit_transform(X)#cross valdition all rows 
lm = LogisticRegression(C=10000)

kf = KFold(n_splits=10, shuffle=True, random_state = 71)
print(np.mean(cross_val_score(lm, X_scaled, y, cv=kf, scoring='accuracy')))
print(np.mean(cross_val_score(lm, X_scaled, y, cv=kf, scoring='f1')))
print(np.mean(cross_val_score(lm, X_scaled, y, cv=kf, scoring='recall')))
# Training MODEL 

lm.fit(X_train_scaled ,y_train)
print("Training Accuracy: ",lm.score(X_train_scaled ,y_train))
#X_test_scaled = scalar.fit_transform(X_test)
y_pre=lm.predict(X_test_scaled)

# orgaize method

y_train_pred = lm.predict(X_train_scaled)# for train accuracy 

print_score(y_train, y_train_pred, train=True)
print_score(y_test, y_pre, train=False)
```

```python
lm = LogisticRegression()

kf = KFold(n_splits=10, shuffle=True, random_state = 71)
print(np.mean(cross_val_score(lm, X_scaled, y, cv=kf, scoring='accuracy')))
print(np.mean(cross_val_score(lm, X_scaled, y, cv=kf, scoring='f1')))
print(np.mean(cross_val_score(lm, X_scaled, y, cv=kf, scoring='recall')))
# Training MODEL 

lm.fit(X_train_scaled ,y_train)
print("Training Accuracy: ",lm.score(X_train_scaled ,y_train))
#X_test_scaled = scalar.fit_transform(X_test)
y_pre=lm.predict(X_test_scaled)

# orgaize method

y_train_pred = lm.predict(X_train_scaled)# for train accuracy 

print_score(y_train, y_train_pred, train=True)
print_score(y_test, y_pre, train=False)
```

### Feature Importance 

```python
import matplotlib.pyplot as pyplot
# get importance
importance = lm.coef_[0]
# summarize feature importance
for i,v in enumerate(importance):
    print('Feature: %0d, Score: %.5f' % (i,v))
# plot feature importance
pyplot.bar([x for x in range(len(importance))], importance)
pyplot.show()
```

### 2-Random Forest (RF)

```python
from sklearn.ensemble import RandomForestClassifier

rf_clf = RandomForestClassifier(n_estimators=100, oob_score=False)# replace 


#kf = KFold(n_splits=10, shuffle=True, random_state = 71)
#print(np.mean(cross_val_score(rf_clf, X_scaled, y, cv=kf, scoring='accuracy')))
#print(np.mean(cross_val_score(rf_clf, X_scaled, y, cv=kf, scoring='f1')))
#print(np.mean(cross_val_score(rf_clf, X_scaled, y, cv=kf, scoring='recall')))

rf_clf.fit(X_train_scaled, y_train)
print("Training Accuracy: ",rf_clf.score(X_train_scaled ,y_train))

y_test_pred = rf_clf.predict(X_test_scaled)# For test accuracy

# make sure #
print("Accuracy",rf_clf.score(X_test_scaled,y_test))
print("Recall ",metrics.recall_score(y_test,y_test_pred))
print("F1_Score",metrics.f1_score(y_test,y_test_pred))

y_train_pred = rf_clf.predict(X_train)# for train accuracy 

print_score(y_train, y_train_pred, train=True)
print_score(y_test, y_test_pred, train=False)

```

### Feature Importance after RF

```python
# get importance
importance = rf_clf.feature_importances_
# summarize feature importance
for i,v in enumerate(importance):
    print('Feature: %0d, Score: %.5f' % (i,v))
# plot feature importance
pyplot.bar([x for x in range(len(importance))], importance)
pyplot.show()
```
